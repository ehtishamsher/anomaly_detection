```python
from sklearn.cluster import DBSCAN
import matplotlib.pyplot as plt
from sklearn.datasets import make_circles
%matplotlib inline
```

```python
X, y = make_circles(n_samples=1000, factor=0.5, noise=0.1)
```

```python
X
```

```python
plt.scatter(X[:, 0], X[:, 1],c=y, )
```

```python
from sklearn.neighbors import LocalOutlierFactor
lof = LocalOutlierFactor(n_neighbors=35,algorithm='ball_tree',leaf_size=30,metric='euclidean',contamination=0.1,novelty=False,n_jobs=-1,p=2)
```

```python
labels = lof.fit_predict(X)
```

```python
print(labels)
```

```python
plt.scatter(X[:, 0], X[:, 1],c = labels)
```

```python
# LOF
labels_lof = lof.fit_predict(X)

colors_lof = ['red' if label == -1 else 'blue' for label in labels_lof]

plt.scatter(X[:, 0], X[:, 1], c=colors_lof, alpha=0.5)
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.title('LOF Anomaly Detection')
plt.show()
```

```python
## verifing 
scores = lof.negative_outlier_factor_

import matplotlib.pyplot as plt
plt.hist(scores, bins=50)
plt.axvline(x=-1.5, color='red', linestyle='--', label='threshold')
plt.xlabel('LOF Score')
plt.ylabel('Count')
plt.title('LOF Score Distribution')
plt.legend()
plt.show()
```
